```python
#riassumendo gli ultimi passaggi della lezione precedente

filePath = "data/serves.csv"
file = open(filePath)
file_letto = file.readlines()

#dividiamo le posizioni di Nadal da quelle di Djokovic
x_nadal = []
y_nadal = []
x_djoko = []
y_djoko = []

###PARSING di un file: lettura tramite interprete che finisce con i dati di interesse
#messi in strutture "comode" da usare
for riga in file_letto: 
    ##possiamo "includere" la prima riga, la ignoriamo con gli "if"
    riga_pulita_splittata = riga.strip().split(',') 
    if riga_pulita_splittata[2] == "Nadal":
        x_nadal.append(float(riga_pulita_splittata[3]))
        y_nadal.append(float(riga_pulita_splittata[-1]))
    elif riga_pulita_splittata[2] == "Djokovic":
        x_djoko.append(float(riga_pulita_splittata[3]))
        y_djoko.append(float(riga_pulita_splittata[-1]))
```

```python
#calcoliamo le medie e plottiamole
def calcola_media(lista_voti):
    somma = 0
    for voto in lista_voti:
        somma = voto + somma
    
    media = somma / len(lista_voti)
    
    return media

import numpy as np
print(calcola_media(x_nadal))
print(np.mean(x_nadal))

nadal_x_medio = np.mean(x_nadal)
nadal_y_medio = np.mean(y_nadal)
djoko_x_medio = np.mean(x_djoko)
djoko_y_medio = np.mean(y_djoko)


plt.scatter(nadal_x_medio, nadal_y_medio)
```

```python
##separiamo le medie per quadrante
"""print(x_nadal)
print(y_nadal)"""
x_limite = 8.23/2
y_limite = 23.77/2

#var1x = '1x'
# var1y = '1y'

##prepariamo DOVE mettere i dati dei singoli quadranti ( in questo caso saranno delle liste
## tutte, per comodità, incluse in un dizionario. Nulla ci vietava di scrivere 4 liste separate)
quadrante = {
    #chiave : valore,
    '1x' : [],
    '1y' : [],
    '2x' : [],
    '2y' : [],
    '3x' : [],
    '3y' : [],
    '4x' : [],
    '4y' : []
}

# quadrante['3x'].append(valore)
#x_nadal[3] mi accede al quarto elemento

for i in range( len(x_nadal) ) :
    #print(i)
    #posso prendere l'elemento "i-esimo"
    #print(x_nadal[i], y_nadal[i])
    
    ##capire in che quadrante sta ogni punto
    if x_nadal[i] < x_limite and y_nadal[i] > y_limite:
        ##questo pezzo di codice verrà eseguito SOLO
        ##se sto nel primo quadrante (alto sx)
        quadrante['1x'].append(x_nadal[i])
        quadrante['1y'].append(y_nadal[i])
    if x_nadal[i] > x_limite and y_nadal[i] > y_limite:
        quadrante['2x'].append(x_nadal[i])
        quadrante['2y'].append(y_nadal[i])
    if x_nadal[i] > x_limite and y_nadal[i] < y_limite:
        quadrante['3x'].append(x_nadal[i])
        quadrante['3y'].append(y_nadal[i])
    if x_nadal[i] < x_limite and y_nadal[i] < y_limite:
        quadrante['4x'].append(x_nadal[i])
        quadrante['4y'].append(y_nadal[i])
    
print(quadrante)
```
